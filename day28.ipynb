```python
#Write a Python program to perform Named Entity Recognition (NER) on a given text using SpaCy. Print the entities and their types.
```

```python
!pip install -U spacy==3.7.1
!python -m spacy download en_core_web_sm
import spacy
nlp = spacy.load("en_core_web_sm")#Loading the English language model
text = "Natural Language processing is one of the trending subject."
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```
